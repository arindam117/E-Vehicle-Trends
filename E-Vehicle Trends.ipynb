```python
 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# THis will help fetch data from google trends
from pytrends.request import TrendReq   

# Creating a pytrend object to request data from google trends
# hl is host language
pytrends = TrendReq(hl='en-us')  


# Extract data about keywords 
keywords = ['electric car','solar car','Hydrogen car','e-bike','petrol car']

# The timeframe = 'today 5-y' gives data back to 5 years from today
pytrends.build_payload(keywords,timeframe = 'today 5-y') 

# This uses the .interest_over_time() module to give a cummulative estimate 

data = pytrends.interest_over_time()


data.plot(kind='line')  

plt.title('Trends in E-Vehicle Searches') # This gives the heading / title of the plot 

plt.xlabel('Year Stamps') # This renames the X-axis label

plt.ylabel('Average Enquiries') # This renames the Y=axis lables

plt.show()
```
